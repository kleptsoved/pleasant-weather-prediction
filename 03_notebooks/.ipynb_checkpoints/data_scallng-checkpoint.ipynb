## 1. Import additional libraries for data scalling

````python
import sklearn
import pandas as pd
from sklearn.preprocessing import StandardScaler
````

````python
for i, element in enumerate(df):
    print(f"Element {i}: type = {type(element)}")
````

````python
df[0].reset_index(inplace=True)
df[0].rename(columns={'index':'id'}, inplace=True)
df[0].head()
````

---
## 2. Scalling
---

### 2.1 Subset Data

````python
df[0].info()
````

### identify the useful numeric columns

````python
columns = df[0].select_dtypes(include=['number']).columns.tolist()
print(f"ðŸ“Š Found {len(columns)} numeric columns:")
print(columns)
````

## 2.2 Check and remove missing values

````python
df[0].shape
````

````python
df[0].isnull().sum()
````

````python
#eliminate those records that have NaN values

df[0].dropna(inplace=True)
df[0].shape
````

### scale all numeric columns, you'll want to exclude identifier columns like 'id' and 'DATE' since those shouldn't be scaled

````python
# Exclude identifier columns that shouldn't be scaled
# Identify columns to exclude (not to be scaled)
excluded_columns = ['id', 'DATE', 'MONTH']
columns_to_scale = [col for col in columns if col not in excluded_columns]
print(f"ðŸ“Š Selected {len(columns_to_scale)} columns for scaling")

# Store the excluded columns separately
df_excluded = df[0][excluded_columns].copy()

# Create subset with only the columns we want to scale
df2 = df[0][columns_to_scale].copy()
df2.head()
````

## 3. Scalling

````python
#create scaler object using StandardScaler from sklearn.preprocessing
#StandardScaler assumes data is normally distributed and scales with a distribution
#around 0 and a standard deviation of 1. Scaling happens independently with each variable.

scaler = StandardScaler()
````

````python
#create new df with scaled data

df_scaled = pd.DataFrame(scaler.fit_transform(df2), columns=columns_to_scale)
df_scaled.head()
````

````python
# Combine excluded columns with the scaled DataFrame
df_final = pd.concat([df_excluded.reset_index(drop=True), df_scaled.reset_index(drop=True)], axis=1)

df_final.head()  # Optional: preview the result
````

````python
# Add this cell at the end of your notebook
print("ðŸ FINALIZING DATA EXPORT")
print("="*50)

# Summary of what you've accomplished
print(f"ðŸ“Š Original dataset: {df[0].shape[0]:,} rows Ã— {df[0].shape[1]} columns")
print(f"âš¡ Scaled dataset: {df_final.shape[0]:,} rows Ã— {df_final.shape[1]} columns")

# Export the final processed data
print(f"\nðŸ“¤ Exporting final scaled dataset...")
exported_file = export_data_interactive(
    df=df_final,
    output_path=output_path,
    original_filename="Dataset-weather-prediction-dataset-processed.csv"
)

print(f"\nðŸŽ‰ Data processing and export completed!")
print(f"ðŸ“ Final file: {exported_file}")
````
